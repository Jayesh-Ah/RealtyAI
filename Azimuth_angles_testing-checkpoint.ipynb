```python
import math
import matplotlib.pyplot as plt
import pytz
import pysolar.solar as ps
import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def angle_between_points(x1, y1, x2, y2):
    delta_x = x2 - x1
    # print('Delta_x:',delta_x)
    delta_y = y2 - y1
    # print("Delta_y:",delta_y)
    angle_rad = math.atan2(delta_y, delta_x)
    # print('Angle_rad:',angle_rad)
    angle_deg = math.degrees(angle_rad)
    # print('Angle_deg:', angle_deg)
    return angle_deg

def get_sun_data(latitude, longitude, date_time):
    # Convert date and time to datetime object
    timezone = pytz.timezone('Asia/Kolkata')

    # # Convert date_time to timezone
    date_time_local = timezone.localize(date_time)

    # # Calculate sun position using pysolar
    sun_elevation = ps.get_altitude(latitude, longitude, date_time_local)
    sun_azimuth = ps.get_azimuth(latitude, longitude, date_time_local)

    return sun_elevation, sun_azimuth

def shadow_output(angle1,azimuth,angle2):
    if angle1<= azimuth <=angle2 or angle1 >= azimuth >= angle2:
        return 'TRUE'
    else:
        return 'FALSE'

def spherical_to_cartesian(elevation, azimuth):
    r = 10000
    x = r * np.sin(np.deg2rad(azimuth)) * np.cos(np.deg2rad(elevation))
    y = r * np.cos(np.deg2rad(azimuth)) * np.cos(np.deg2rad(elevation))
    z = r * np.sin(np.deg2rad(elevation))
    return x, y, z

latitude = 28.4595
longitude = 77.0266

time = datetime.datetime(2023, 7, 15, 14, 0, 0)

elevation,azimuth=get_sun_data(latitude,longitude,time)
print("Sun Elevation: ", elevation," degrees")
print("Sun Azimuth: ", azimuth," degrees")

elevation_rad, azimuth_rad = np.deg2rad(elevation), np.deg2rad(azimuth)
# sun_x, sun_y, sun_z = spherical_to_cartesian(elevation_rad, azimuth_rad)
# Example usage:
x1, y1 = 790,-695   #Observer Wall Point 1
x2, y2 = 510,-440   #Obstacle Wall Point 1
x3 ,y3 = 510,-1000      #Obstacle Wall Point 2
angle_2 = angle_between_points(x1, y1, x2, y2)
if angle_2 > 90 and angle_2 < 180:
    angle_2n= (450-angle_2) % 360
else:
    angle_2n=90-angle_2

angle_3=angle_between_points(x1, y1, x3, y3)
if angle_3 > 90 and angle_3 < 180:
    angle_3n= (450-angle_3) % 360
else:
    angle_3n=90-angle_3
print("Angle between (x1,y1) and (x2,y2) i.e Angle2n:", angle_2n, "degrees")
print("Angle between (x1,y1) and (x3,y3): i.e angle3n", angle_3n, "degrees")
# print("Angle between (x1,y1) and (x2,y2) i.e Angle2:", angle_2, "degrees")
# print("Angle between (x1,y1) and (x3,y3): i.e angle3", angle_3, "degrees")

shadow_status=shadow_output(angle_2n,azimuth,angle_3n)
print('angle_3n:',angle_3n)
print('angle_2n:',angle_2n)
print('Time: ', time)
print("Shadow Azimuth:-",shadow_status)



#Plotting
sun_x, sun_y, _ = spherical_to_cartesian(elevation, azimuth)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot([x1, x2], [y1, y2], 'k', label=f'Angle 2: {angle_2n:.2f}°')
plt.plot([x1, x3], [y1, y3], 'r-', label=f'Angle 3: {angle_3n:.2f}°')
plt.plot([x2, x3], [y2, y3], 'b-', label='Building Wall')
# plt.plot(x1,azimuth,label='Azimuth angle')
plt.plot(x1, y1, 'go', label='Point (x1, y1)')
plt.plot(x2, y2, 'bo', label='Point (x2, y2)')
plt.plot(x3, y3, 'ro', label='Point (x3, y3)')
plt.quiver(x1, y1, sun_x - x1, sun_y - y1, scale=0.2,color='orange', label='Sun Azimuth')
plt.xlabel('West-East')
plt.ylabel('South-North')
plt.title('Angle between Points')
plt.legend()
plt.suptitle(f" Azimuth: {azimuth}")

plt.grid(True)
plt.show()
```

```python
import os
import datetime
import numpy as np
import matplotlib.pyplot as plt
import pytz
from pysolar.solar import get_altitude, get_azimuth

def calculate_cross_product(observer_x, observer_y, obstacle_x, obstacle_y):
    # Cross product of vectors OA (observer to obstacle) and unit vector in direction of sun
    return (obstacle_x - observer_x) * 1  - (obstacle_y - observer_y) * 0  # Unit vector [1,0] assumed for simplicity

# Function to calculate azimuth with respect to positive Y axis in clockwise direction
def calculate_azimuth(x_observer, y_observer, x_target, y_target, is_ccw=False):
    if (x_target == x_observer):
        return 0
    angle = np.degrees(np.arctan2(y_target - y_observer, x_target - x_observer)) % 360
    if is_ccw:
        return -angle + 90  # Negative for counterclockwise
    return 90 - angle

# Wall orientation determination function
def wall_orientation(ab_aw_z):
    if ab_aw_z < 0:
        return "Front"
    elif ab_aw_z > 0:
        return "Back"
    else:
        return "Ofob"

# Function to get solar data (elevation and azimuth)
def get_sun_data(latitude, longitude, date_time):
    timezone = pytz.timezone('Asia/Kolkata')  # Use 'Asia/Kolkata' as a placeholder, adjust as needed
    date_time_local = timezone.localize(date_time)
    sun_elevation = get_altitude(latitude, longitude, date_time_local)
    sun_azimuth = get_azimuth(latitude, longitude, date_time_local)
    return sun_elevation, sun_azimuth

# Function to calculate cross products and orientations of obstacles with respect to observers
def calculate_obstacle_orientations(obstacle_coords, observer_coords):
    orientations = {}

    # Loop through each observer's coordinates
    for idx, (x_obs, y_obs) in enumerate(observer_coords):
        orientations[idx] = []  # Initialize a list for this observer's orientations
        # Loop through each obstacle point
        for obs_x, obs_y in obstacle_coords:
            # Calculate the cross product for the observer and the obstacle
            cross_product = calculate_cross_product(x_obs, y_obs, obs_x, obs_y)
            # Determine the orientation of the obstacle wall relative to the observer
            orientation = wall_orientation(cross_product)
            orientations[idx].append(orientation)  # Store the orientation for this observer
    
    return orientations  # Return the dictionary containing orientations for all observers

# Generalized function to handle azimuth calculation and plotting for both observer points in the same plot
def handle_observer_obstacle(observer_coords, obstacle_coords, latitude, longitude, time_range, save_dir):
    os.makedirs(save_dir, exist_ok=True)

    # Calculate orientations for the obstacles with respect to the observers
    obstacle_orientations = calculate_obstacle_orientations(obstacle_coords, observer_coords)

    for time in time_range:
        # Get sun data
        elevation, azimuth = get_sun_data(latitude, longitude, time)
        if elevation <= 0:
            print(f"Skipping plot for {time.strftime('%H:%M')}, as sun is below the horizon.")
            continue

        # Initialize shadow flags for both observers
        shadow_flag_1 = False
        shadow_flag_2 = False

        # Plotting
        plt.figure(figsize=(12, 10))

        for idx, (x_obs, y_obs) in enumerate(observer_coords):
            # Cross product calculations for each observer
            ab_aw1_z = calculate_cross_product(x_obs, y_obs, obstacle_coords[0][0], obstacle_coords[0][1])
            ab_aw2_z = calculate_cross_product(x_obs, y_obs, obstacle_coords[1][0], obstacle_coords[1][1])
            print(f'ab_aw1_z: {ab_aw1_z}')
            print(f'ab_aw2_z: {ab_aw2_z}')

            # Retrieve orientation for the current observer regarding both obstacles
            orientation_1 = obstacle_orientations[idx][0]
            orientation_2 = obstacle_orientations[idx][1]
            
            print(f"Time: {time.strftime('%H:%M')} -> Orientation of obstacle wall wrt Observer {idx+1}: {orientation_1}")
            print(f"Time: {time.strftime('%H:%M')} -> Orientation of obstacle wall wrt Observer {idx+1}: {orientation_2}")

            # Logic to calculate azimuth for front and back cases
            azimuth_point1 = azimuth_point2 = None
            if (ab_aw1_z * ab_aw2_z < 0):
                azimuth_point1 = calculate_azimuth(x_obs, y_obs, obstacle_coords[0][0], obstacle_coords[0][1])
                azimuth_point2 = calculate_azimuth(x_obs, y_obs, obstacle_coords[1][0], obstacle_coords[1][1], is_ccw=True)
            else:  
                azimuth_point2 = calculate_azimuth(x_obs, y_obs, obstacle_coords[1][0], obstacle_coords[1][1])%360
                azimuth_point1 = calculate_azimuth(x_obs, y_obs, obstacle_coords[0][0], obstacle_coords[0][1])%360

            print(f"Observer {idx+1} -> Azimuth of Point 1 : {azimuth_point1:.2f}°")
            print(f"Observer {idx+1} -> Azimuth of Point 2 : {azimuth_point2:.2f}°")

            # Sun's Cartesian coordinates for quiver plot (each observer)
            sun_x = np.cos(np.radians(elevation)) * np.sin(np.radians(azimuth))
            sun_y = np.cos(np.radians(elevation)) * np.cos(np.radians(azimuth))

            # Determine shadow flags based on azimuth and wall orientation
            if azimuth_point1 <= azimuth <= azimuth_point2 or azimuth_point1 >= azimuth >= azimuth_point2:
                shadow_flag_1 = True if idx == 0 else shadow_flag_1
                shadow_flag_2 = True if idx == 1 else shadow_flag_2

            # Plot lines for each observer
            plt.plot([x_obs, obstacle_coords[0][0]], [y_obs, obstacle_coords[0][1]], 'k-',
                     label=f'Obs {idx+1}: Angle to Point 1 (Azimuth: {azimuth_point1:.2f}° CW)')
            plt.plot([x_obs, obstacle_coords[1][0]], [y_obs, obstacle_coords[1][1]], 'r-',
                     label=f'Obs {idx+1}: Angle to Point 2 (Azimuth: {azimuth_point2:.2f}° CCW)')

            # Plot obstacle wall line
            plt.plot([obstacle_coords[0][0], obstacle_coords[1][0]], [obstacle_coords[0][1], obstacle_coords[1][1]], 
                     'b--', label='Obstacle Wall')

            # Plot observer and obstacle points
            plt.plot(x_obs, y_obs, 'go', label=f'Observer {idx+1}')
            plt.plot(obstacle_coords[0][0], obstacle_coords[0][1], 'ro', label='Obstacle Point 1')
            plt.plot(obstacle_coords[1][0], obstacle_coords[1][1], 'ro', label='Obstacle Point 2')

            # Azimuth arrow representing the sun's azimuth angle from the observer point
            plt.quiver(x_obs, y_obs, np.sin(np.radians(azimuth)), np.cos(np.radians(azimuth)),
                       scale=5, color='orange', label=f'Sun Azimuth (Obs {idx+1}: {azimuth:.2f}°)', angles='xy')

        # Plot settings
        plt.xlabel('West-East')
        plt.ylabel('South-North')
        plt.title(f'Sun Azimuth: {azimuth:.2f}° at {time.strftime("%H:%M")}')

        # Shadow flag status
        shadow_status = shadow_flag_1 or shadow_flag_2
        plt.text(0.05, 0.95, f'Shadow Flag Status: {"True" if shadow_status else "False"}',
                 transform=plt.gca().transAxes, fontsize=12, color='red')

        plt.grid(True)
        plt.legend()
        plt.axis('equal')

        # Save plot
        plot_filename = f'{time.strftime("%H_%M")}_Combined_Observers_plot.png'
        plt.savefig(os.path.join(save_dir, plot_filename))
        plt.close()
```
